```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder=LabelEncoder()
```

```python
for col_name in df1.columns:
  if df1[col_name].dtype=="object":
    df1[col_name]=encoder.fit_transform(df1[[col_name]])
```

```python
scaler=MinMaxScaler()
```
